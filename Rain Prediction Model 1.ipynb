```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/LearnPythonWithRune/MachineLearningWithPython/main/jupyter/final/files/weather.csv',parse_dates=True,index_col=0)
```

```python
data.head()
```

```python
data.index
```

```python
datax= data[['Humidity3pm','Pressure3pm','RainTomorrow']]
datax
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots()

datax[datax['RainTomorrow'] == 'No'].plot.scatter(x='Humidity3pm', y='Pressure3pm', c='b', alpha=.25, ax=ax)
datax[datax['RainTomorrow'] == 'Yes'].plot.scatter(x='Humidity3pm', y='Pressure3pm', c='r', alpha=.25, ax=ax)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
datax_clean = datax.dropna()
```

```python
x=datax_clean[['Humidity3pm','Pressure3pm']]
y=datax_clean['RainTomorrow']
y=np.array([0 if value == 'No' else 1 for value in y])
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42)
```

```python
neigh = KNeighborsClassifier()
neigh.fit(x_train,y_train)
y_pred=neigh.predict(x_test)
accuracy_score(y_pred,y_test)
```

```python
x_map = np.random.rand(10000,2)
x_map = x_map*(100,50)+(0,990)
```

```python
fig, ax = plt.subplots()

y_map = neigh.predict(x_map)

ax.scatter(x=x_map[:,0], y=x_map[:,1], c=y_map, alpha=.25)
```

```python
scores = []
for k in range(1, 100):
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(x_train, y_train)
    y_pred = neigh.predict(x_test)
    score = accuracy_score(y_test, y_pred)
    scores.append(score)
```

```python
fig,ax =plt.subplots()

ax.plot(scores)
```
